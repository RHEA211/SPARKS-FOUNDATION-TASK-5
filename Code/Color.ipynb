#### Return the hex value of a color given

```python
def RGB2HEX(color):
    return "#{:02x}{:02x}{:02x}".format(int(color[0]), int(color[1]), int(color[2]))
#We have multiplied 2 dimensions of the image to feed Kmeans with a flattened array
```

#### Read and return the image using its path

```python
def get_image(image_path):
    return plt.imread(image_path)
```

## How is colour identification done

**Algorithm used - Kmeans clustering**
- The image is retrieved
- The dimensions are multiplies to get a flat array
- Apply KMeans and then predict on the image to get the results.
- Then, colors are identified in clusters and arranged in the correct order.




```python
def get_colors(image_path, number_of_colors, show_chart):
    fig, (ax1, ax2) = plt.subplots(1, 2)
    image = get_image(image_path)
    ax1.imshow(image)
    
#We use Kmeans algorithm  
modified_image = image.reshape(image.shape[0]*image.shape[1], 3)
    clf = KMeans(n_clusters = number_of_colors)
    labels = clf.fit_predict(modified_image)
    
#The colours are ordered, first hex then RGB colours saved    
    counts = Counter(labels)
    center_colors = clf.cluster_centers_
    ordered_colors = [center_colors[i] for i in counts.keys()]
    hex_colors = [RGB2HEX(ordered_colors[i]) for i in counts.keys()]
    rgb_colors = [ordered_colors[i] for i in counts.keys()]

# Plotting the colours as  pie charts   
if (show_chart):
        ax2.pie(counts.values(), labels = hex_colors, colors = hex_colors)
    
    return rgb_colors
```
